#### To read an h5 file in python using h5py library

```python
import h5py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "beams3d_ncsx_c09r00_free_birth.h5"
data = h5py.File(filename, 'r')
```

```python
# read raxis, zaxis, phiaxis
raxis = data['/raxis'][()]
zaxis = data['/zaxis'][()]
phiaxis = data['/phiaxis'][()]
```

```python
list(data.keys())
```

```python
# read b_phi
b_phi = data['/B_PHI'][()]
b_r = data['/B_R'][()]
b_z = data['/B_Z'][()]
```

```python
# plot b_phi, b_r, b_z
ax, fig = plt.subplots(1, 3, figsize=(15, 5))
fig[0].contourf(raxis, zaxis, b_phi[:, 0, :], 100, cmap='jet')
fig[0].set_title('$B_{\phi}$')
fig[0].set_xlabel('$r$')
fig[0].set_ylabel('z')

fig[1].contourf(raxis, zaxis, b_r[:, 0, :], 100, cmap='jet')
fig[1].set_title('$B_r$')
fig[1].set_xlabel('$r$')

fig[2].contourf(raxis, zaxis, b_z[:, 0, :], 100, cmap='jet')
fig[2].set_title('$B_z$')
fig[2].set_xlabel('$r$')

# colorbar at the bottom for all subplots spanning the whole width
plt.colorbar(fig[0].contourf(raxis, zaxis, b_phi[:, 0, :], 100, cmap='jet'), ax=fig, orientation='horizontal', fraction=0.05, pad=0.1)

# save the plot
plt.savefig('beams3d_ncsx_c09r00_free_birth.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# read wall_vertex
wall_vertex = data['/wall_vertex'][()]
```

```python
# plot wall_vertex using raxis, zaxis
plt.figure()
plt.plot(wall_vertex[:, 0], wall_vertex[:, 2])
plt.title('Wall Vertex')
plt.xlabel('R')
plt.ylabel('Z')
plt.show()
```
